## Merge

Many a times you will be working with multiple dataframes all at once.

The merge function allows them to be combined into a single data frame

````python
pd.merge(superheroes, 
         publishers, 
         on='publisher', 
         how='left')
````

````python
pd.merge(superheroes, 
         publishers, 
         on='publisher', 
         how='inner')
````

````python
pd.merge(superheroes, 
         publishers, 
         on='publisher', 
         how='right')
````

````python
pd.merge(superheroes1, 
         publishers, 
         on='publisher', 
         how='outer')
````
