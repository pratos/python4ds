### Nicer bigger and clearer plot

### But there are lot of things missing, we'll try to answer a few:

1. Looking just at the plot without the code, I don't know what the plot is about?
2. What does the x-axis and y-axis mean?
3. What color represents what?

### 1. Looking just at the plot without the code, I don't know what the plot is about?

To answer it, `matplotlib` provides way to put titles to a plot, how? let's find out:

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.show()
```

### 2. What does the x-axis and y-axis mean?

To answer it, `matplotlib` provides `xlabels` and `ylabels`

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.show()
```

### Q: Can you try out, `ylabels`?

### 3. What color represents what?

We can try to answer this using `legends`

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
plt.show()
```

### There can be more cleanups to the above plot

* Can remove the ticks on `x-axes` (we don't need number of steps here and also `y-axes` too can be removed)

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
plt.xticks([])
plt.show()
```

* We can add grids at the background

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
# plt.xticks([])
plt.grid(axis='both')
plt.show()
```

### We can save this plot as well!

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
# plt.xticks([])
plt.grid(axis='both')
plt.show()
plt.savefig('../data/python_process.jpg')
```

## Calculating quartlies

* Put the list of numbers in order
* Then cut the list into four equal parts
* The Quartiles are at the "cuts"

```python
list_ = [2, 4, 4, 5, 6, 7, 8]
```

>Percentiles divide the whole population into 100 groups where as quartiles divide the population into 4 groups

* p = 25: First Quartile or Lower quartile (LQ)
* p = 50: second quartile or Median
* p = 75: Third Quartile or Upper quartile (UQ)

```python
np.percentile(np.array(list_), [25, 50, 75, 100])
```
